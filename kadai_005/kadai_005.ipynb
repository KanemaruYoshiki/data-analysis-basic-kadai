```python
import matplotlib.pyplot as plt
import numpy as np
# 折れ線グラフ
fig, axes = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

# サインカーブ
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 10),
         ylim=(-1, 5), yticks=np.arange(-1, 6))

axes[0, 1].plot(x1, y1, linewidth=2.0)

# ヒストグラム
x2 = np.random.normal(15, 5, 2000)
axes[0, 2].set(yticks=np.arange(0, 600, 200))
axes[0, 2].hist(x2)

# 散布図
np.random.seed(1)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, 60)
axes[1, 0].set(ylim=(-2.5, 9), yticks=np.arange(0.0, 9.0, 2.5))
axes[1, 0].scatter(x3, y3)

plt.show()
```
